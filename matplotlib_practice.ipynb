```python
# Import dependensies
import matplotlib.pyplot as plt
#This is the same as typing from matplotlib import pyplot
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
plt.xlim("Jan","June")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.figure(figsize=(2, .5))

# Set the y limit between 0 and 45.
plt.ylim(0, 45)
#plt.xlim(0,5)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
fig, ax = plt.subplots()
#alpha is to make it 20%(1-0.2) transparent
ax.scatter(y_axis, x_axis, color = 'skyblue' , s=[i*5 for i in y_axis] , label = " Boston", alpha=0.8 , edgecolor='black')
ax.invert_yaxis()
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
plt.xlim(0,50)
# Create a title.
ax.set_title("PyBer Fare by Month")
# Add the legend.
ax.legend()
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python

# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0) # which wedge should be seperated
# 0.2 is the fraction of the radius from the center of the pie for each wedge 1 is the max value a wedge can have

plt.subplots(figsize=(8, 8))
# adding ; end of plt.pie will get rid of all the extra statements before showing pie chart
#plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%');
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%');

```
